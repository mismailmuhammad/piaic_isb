```python
#Assignment .................Make use of different string functions in Python.............nt 

Student="Muhammad Ismail"
FatherName="Abdul Karim"
Course="Artificial Intelligence"
Message1="How are you?\tWoud you like to participate in\n Python Quiz."
Message2="ARE you COMMITTED TO learn Machine Learning?"
Message3= "AI COULD PLAY A VITAL ROLE IN THE ECONOMIC DEVELOPMENT OF A COUNTRY?   DO YOU AGREE?"
Message4="Make Sure your Assignment must be uploaded on Github in a timely manner"


# 1-   count number of words repeated in a string
count=Message1.count("you")
print("Ans 1: The count of the given word in Message1=" + str(count))
print("\n")

# 2-   Capitalize  Converts the first character to upper case
print("Ans 2: Student Name in capitalize format=" +Student.capitalize())
print("\n")

# 3-   casefold()	Converts string into lower case
print("Ans 3: Student Name in lowercase="+Student.casefold())
print("\n")

#  4-  center()	Returns a centered string
print("Ans 4: Using Center function\n")
print(Message3.center(100,'*'))
print("\n")


#  5-  encode()	Returns an encoded version of the string
encoded=Message1.encode()
print(" Ans 5: Encoded version of the string:\n")
print(encoded)
print("\n")

#  6-  endswith()	Returns true if the string ends with the specified value
word=Message1.endswith(".")
print(" Ans 6: The Message1 Ends with dot(.)\n")
print(word)
print("\n")

#  7-  expandtabs()	Sets the tab size of the string
tabs=Message1.expandtabs(5)
print(" Ans 7: Use of tabs in a given Messge\n"+tabs)
print("\n")

#   8-  find()	Searches the string for a specified value and returns the position of where it was found
Index=Message1.find("are")
print(" Ans 8 :Index of the specified value="+ str(Index))
print("\n")

#   9-  Format 
txt="Only on {price:.2f} dollars!"
print("Ans 9: Format function\n"+txt.format(price=100))
print("\n")

#   10-  index()	Searches the string for a specified value and returns the position of where it was found
print("Ans 10: The word COMMITTED is at index=",Message2.index("COMMITTED"))
print("\n")

#   11-  isalnum()	Returns True if all characters in the string are alphanumeric
print("Ans 11: Variable Message1 is alphnumeric:\n",Message1.isalnum())
print("\n")

#   12- isalpha()	Returns True if all characters in the string are in the alphabet
print("Ans 12:  Variable Message1 contains all alphabets:\n",Message1.isalpha())
print("\n")

#   13- isdecimal()	Returns True if all characters in the string are decimals
print("Ans 13:  Variable Name contains all decimals:\n",Message1.isdecimal())
print("\n")

#    14-  isidentifier()	Returns True if the string is an identifier
print("Ans 14: Name variable is identifier:\n",Message1.isidentifier())
print("\n")

#  15-  islower()	Returns True if all characters in the string are lower case
print("Ans 15: islower function use\n")
print(Message1.islower())
print("\n")

#  16- isnumeric()	Returns True if all characters in the string are numeric
print("Ans 16: isnumeric function use\n")
print(Message1.isnumeric())
print("\n")

#   17- isprintable()	Returns True if all characters in the string are printable
print("Ans 17: isprintable function use\n")
print(Message1.isprintable())
print("\n")

#   18- isspace()	Returns True if all characters in the string are whitespaces
print("Ans 18: islower function use\n")
print(Message1.isspace())
print("\n")
 
#  19- istitle() 	Returns True if the string follows the rules of a title
print("Ans 19: istitle function use\n")
print(Message1.istitle())
print("\n")

#  20-  isupper()	Returns True if all characters in the string are upper case
print("Ans 20: isupper function use\n")
print(Message1.isupper())
print("\n")

#  21-  ljust()	Returns a left justified version of the string
x='AI can make your future.'
print("Ans 21: ljust function use\n")
print(x.ljust(30,'*'))
print("\n")

#  22-  lower()	Converts a string into lower case
print("Ans 22: lower function use\n")
print("lower case string....\n",Message4.lower())
print("\n")

#  23- lstrip()	Returns a left trim version of the string
print("Ans 23: lstrip function use\n")
print("lstrip method\n",Message4.lstrip())
print("\n")

#  24- partition()	Returns a tuple where the string is parted into three parts
print("Ans 24: partition function use\n")
print("Partition function result :",Message2.partition("learn"))
print("\n")

#  25-  replace()	Returns a string where a specified value is replaced with a specified value
print("Ans 25: replace function use\n")
print(Message2.replace("COMMITTED","WILLING"))
print("\n")

#  26-  rfind()	Searches the string for a specified value and returns the last position of where it was found
print("Ans 26: rfind function use\n")
print("Last position of word you\n",Message1.rfind("you"))
print("\n")

#  27- rindex()	Searches the string for a specified value and returns the last position of where it was found
print("Ans 27: rindex function use\n")
print("Last index of word you\n",Message1.rindex("you"))
print("\n")

#  28- rjust()	Returns a right justified version of the string
y='Commitment and Capacity leads to implementation'
print("Ans 28: rjust function use\n")
print(y.rjust(60,'*'))
print("\n")

#  29- rpartition()	Returns a tuple where the string is parted into three parts
print("Ans 29: rsplit function use\n")
print(Message4.rsplit())
print("\n")

#  30-- rsplit()	Splits the string at the specified separator, and returns a list
print("Ans 30: splitlines function use\n")
print(Message4.splitlines())
print("\n")

#  31- rstrip()	Returns a right trim version of the string
print("Ans 31: rstrip function use\n")
w='aaaaaaaaHow Could I Remove the leading asaaaaaaaaaaaaaaaa'
print(w.rstrip('a'))
print("\n")

#  32- splitlines()	Splits the string at line breaks and returns a list
print("Ans 32: splitlines function use\n")
print(Message4.splitlines())
print("\n")


#  33-  startswith()	Returns true if the string starts with the specified value
print("Ans 33: islower function use\n")
print(Message1.startswith("How"))
print("\n")

#  34-  strip()	Returns a trimmed version of the string
print("Ans 34: strip function use\n")
z='aaaaaaaaHow Could I Remove the leading as'
print(z.strip('a'))
print("\n")
 
#  35- swapcase()	Swaps cases, lower case becomes upper case and vice versa
print("Ans 35: swapcase function use\n")
print(Message4.swapcase())
print("\n")
 
#  36- title()	Converts the first character of each word to upper case
print("Ans 36: title function use\n")
print(Message4.title())
print("\n")
 
#  37-  upper()	Converts a string into upper case
print("Ans 37: upper function use\n")
print(Message4.upper())
print("\n")
 
#  38-  zfill()	Fills the string with a specified number of 0 values at the beginning
msg="Good Bye"
print("Ans 38: zfill function use\n")
print(msg.zfill(70))
print("\n")


```
